```python
df['Lowercase_Pandas'] = df['reviews.text'].str.lower()#panda toLower()
```

```python
nlp = sp.load("en_core_web_sm")
```

```python
df['Lowercase_spaCy'] = df['reviews.text'].apply(lambda x: nlp(x).text.lower())#spacy Lowercase
```

```python
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('punkt_tab')
nltk.download('stopwords')
```

```python
#stopwords nltk si spacy
stop_words_nltk = set(stopwords.words('english'))
stop_words_spacy = set([w.lower() for w in nlp.Defaults.stop_words])
```

```python
df["tokens_pandas"] = df["Lowercase_Pandas"].apply(word_tokenize)
```

```python
df["no_stop_nltk"] = df["tokens_pandas"].apply(
    lambda tokens: [t for t in tokens if t not in stop_words_nltk]
)

df["no_stop_spacy"] = df["tokens_pandas"].apply(
    lambda tokens: [t for t in tokens if t not in stop_words_spacy]
)
```

```python
df.head()
```

```python
import numpy as np
numpy_nltk = np.array(df["no_stop_nltk"].tolist(), dtype=object)
numpy_spacy = np.array(df["no_stop_spacy"].tolist(), dtype=object)

```

```python
print(len(stop_words_nltk), len(stop_words_spacy))
```

```python

def remove_stopwords_nltk(text):
    tokens = word_tokenize(text)
    filtered = [t for t in tokens if t not in stop_words_nltk]
    return " ".join(filtered)

def remove_stopwords_spacy(text):
    doc = nlp(text)
    filtered = [token.text for token in doc if token.text.lower() not in stop_words_spacy]
    return " ".join(filtered)


df["No_Stopwords_NLTK"] = df["Lowercase_Pandas"].apply(remove_stopwords_nltk)
df["No_Stopwords_SPACY"] = df["Lowercase_Pandas"].apply(remove_stopwords_spacy)
```

```python
df.head()
```

```python
df["len_no_stop_nltk"] = df["No_Stopwords_NLTK"].apply(lambda x: len(x.split()))

#numarare tokeni
df["len_no_stop_spacy"] = df["No_Stopwords_SPACY"].apply(lambda x: len(x.split()))

df["diff_nltk_spacy"] = df["len_no_stop_nltk"] - df["len_no_stop_spacy"]
```
