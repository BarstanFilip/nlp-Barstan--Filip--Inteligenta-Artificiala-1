```python
import pandas as pd
```

```python
df = pd.read_csv('date.csv')
```

```python
df = df.drop(df.columns[[3,5,6,8,9,10,11]], axis =1)
```

```python
df = df.rename(columns={'tokens_pandas': 'Tokenized_NLTK'})
```

```python
df.head()
```

```python
from sklearn.preprocessing import MultiLabelBinarizer
mlb = MultiLabelBinarizer()
df_onehot_nlk = pd.DataFrame(mlb.fit_transform(df['Tokenized_NLTK']),columns=mlb.classes_, index=df.index)
```
